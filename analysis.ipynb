```python
import pandas as pd
import numpy as np
import category_encoders as ce
```

```python
csv_path = "/Users/ilnarkhasanov/WorkSpace/Study/LSD/collected_metrics/ClickHouse-Native-JDBC/class.csv"
csv = pd.read_csv(csv_path)
csv.drop(["file", "class"], axis=1, inplace=True)

categorical_columns = ["type"]
encoder = ce.OneHotEncoder(cols=categorical_columns, use_cat_names=True)
csv_encoded = encoder.fit_transform(csv)
```

```python
csv_encoded
```

```python
vectors = csv_encoded.to_numpy()
```

```python
vectors
```

```python
correlation_matrix = np.corrcoef(vectors, rowvar=False)
```

```python
correlation_matrix
```

```python
cm_df = pd.DataFrame(correlation_matrix)
cm_df.fillna(cm_df.mean(), inplace=True)
```

```python
cm_df
```

```python
flattened_corr = correlation_matrix.flatten()
flattened_corr_ids = []

added_pairs = []

for i in range(len(flattened_corr)):
    pair = (i // 54, i % 54)

    if pair[0] == pair[1] or flattened_corr[i] == 1 or np.isnan(flattened_corr[i]) or (pair[1], pair[0]) in added_pairs:
        continue

    added_pairs.append(pair)

    flattened_corr_ids.append((pair, flattened_corr[i]))
flattened_corr_ids.sort(key=lambda x: x[1], reverse=True)
```

```python
flattened_corr_ids
```

```python
columns = csv_encoded.columns
```

```python
# 30 the most correlated ones
for (row, column), correlation_coefficient in flattened_corr_ids[:30]:
    print(columns[row], columns[column], correlation_coefficient)
```
